# ‚úîÔ∏è Parte 3 - Columnas calculadas

Funci√≥n 1: Categorizaci√≥n de BMI

```python
# Funci√≥n categor√≠a seg√∫n BMI
def cat_bmi(bmi):
  try:
    if bmi >= 0 and bmi <20:
      return 'Bajo'
    elif bmi >= 20 and bmi < 40:
      return 'Medio'
    elif bmi >= 40:
      return 'Alto'
  except TypeError:
    print('Error en el tipo de dato')
```

```python
# Creo la nueva columna Categor√≠a BMI
df['Categor√≠a BMI'] = df[['bmi']].applymap(cat_bmi)
```

Funci√≥n 2: Categorizaci√≥n por Nivel Promedio de Glucosa

```python
# Funci√≥n nivel de glucosa
def avg_glucose_level(glucose):
  try:
    if glucose >= 0 and glucose < 100:
      return 'Bajo'
    elif glucose >= 100 and glucose < 200:
      return 'Medio'
    elif glucose >= 200:
      return 'Alto'
  except TypeError:
    print('Error en el tipo de dato')
```

```python
# Creo la nueva columns de Categor√≠a Nivel de Glucosa
df['Categor√≠a Nivel de Glucosa'] = df[['avg_glucose_level']].applymap(avg_glucose_level)
```

# ‚úîÔ∏è Parte 6 - SQL

Para complementar el an√°lisis, realizar las siguientes consultas SQL

- Obtener todos los registros donde la persona est√° casada y no casada.


```python
# Consulta SQL
query = """
SELECT *
FROM df
GROUP BY ever_married
"""

# Ejecutar la consulta
result = psql.sqldf(query, )

# Mostrar el resultado
print('Datos de personas casadas y no casadas:')
print(result)
```

 - Obtener todos los registros de personas mayores a 20 a√±os


```python
#Consulta SQL
query = """
SELECT *
FROM df
WHERE age > 20
"""

# Ejecutar la consulta
result = psql.sqldf(query, globals())

# Mostrar el resultado
print('Personas mayores a 20 a√±os:')
print(result)
```

- Obtener todos los registros de personas con hipertensi√≥n y sin hipertensi√≥n.


```python
query = """
SELECT *
FROM df
GROUP BY hypertension
"""

# Ejecutar la consulta
result = psql.sqldf(query, globals())

# Mostrar el resultado
print('Personas con y sin hipertensi√≥n:')
print(result)
```

 - Obtener todos los registros de personas que tuvieron stroke y las que no tuvieron

```python
query = """
SELECT *
FROM df
GROUP BY stroke
"""

# Ejecutar la consulta
result = psql.sqldf(query, globals())

# Mostrar el resultado
print('Personas con y sin stroke:')
print(result)
```

- Obtener todos los registros de personas que tuvieron stroke y son menores a 10 a√±os

```python
query = """
SELECT *
FROM df
WHERE age < 10 AND stroke = 1
"""

# Ejecutar la consulta
result = psql.sqldf(query, globals())

# Mostrar el resultado
print('Personas menores a 10 a√±os con stroke:')
print(result)
```

- Obtener la suma total de registros y validar que se condise con el `shape` del dataframe (Para esto, podes usar la formula `sum()` indicando la columna `id` entre los par√©ntesis.

```python
query = """
SELECT COUNT(*) as total_registros
FROM df
"""

# Ejecutar consulta
total_registros_sql = psql.sqldf(query, globals())

# Obtener el n√∫mero total de registros
total_registros_sql = total_registros_sql['total_registros'].iloc[0]

# Validar que el n√∫mero total de registros coincide con el shape del DataFrame
total_registros_shape = df.shape[0]
print(f"N√∫mero total de registros (SQL): {total_registros_sql}")
print(f"N√∫mero de registros (shape): {total_registros_shape}")

# Validaci√≥n
if total_registros_sql == total_registros_shape:
    print("El n√∫mero total de registros coincide con el n√∫mero de registros del DataFrame.")
else:
    print("El n√∫mero total de registros no coincide con el n√∫mero de registros del DataFrame.")
```
